```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
outliers.sum()
```

```python
df["outlier_flag"] = outliers.any(axis=1)
df["outlier_flag"].value_counts()
```

```python
df_capped = df.copy()

for col in df.select_dtypes(include="number").columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_capped[col] = df[col].clip(lower, upper)
```

```python
df_capped.corr()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df_capped.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
df_capped.to_csv("cleaned_student_performance.csv", index=False)
```
